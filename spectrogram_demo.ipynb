```python
import warnings
warnings.simplefilter('ignore')
import numpy, scipy.sparse, librosa, CDL
import numpy.fft
```

```python
def get_spectrogram_samples(filename, mels=64, sample_width=8, pad=False):
    # Load the audio data
    (y, sr) = librosa.load(filename, sr=22050)
    
    # Build mel spectrogram
    # hop=512 => 23ms @ 22050Hz
    S       = librosa.feature.melspectrogram(y, sr, n_mels=mels, n_fft=2048, hop_length=256, fmax=8000)
    
    # Normalize
    S       = S / S.max()
    
    N       = S.shape[1] / sample_width
    
    # Truncate any dangling partial frames
    S       = S[:, :(sample_width * N)]
    
    # Reshape into frames
    S       = S.reshape( (S.shape[0], sample_width, N), order='F')
    
    # Build the output array
    if pad:
        xshape  = (2 * mels, 2 * sample_width)
    else:
        xshape  = (mels, sample_width)
        pass
    
    X = numpy.fft.fft2(S, s=xshape, axes=(0,1))
    
    return CDL.complex_to_real2(X.reshape( (X.shape[0] * X.shape[1], X.shape[2]), order='F'))
    
```

```python
def show_2d_basis(D, sample_width=8, pad=False):
    # First, squash back to columns
    Dcol   = CDL.real2_to_complex(CDL.diags_to_columns(D))
    
    if pad:
        Dcol   = Dcol.reshape( (Dcol.shape[0] / (2*sample_width), 2*sample_width, Dcol.shape[1]), order='F')
        Dtime  = numpy.empty( (Dcol.shape[0] / 2, sample_width, Dcol.shape[1]), order='F')
        pass
    else:
        Dcol   = Dcol.reshape( (Dcol.shape[0] / sample_width, sample_width, Dcol.shape[1]), order='F')
        Dtime  = numpy.empty_like(Dcol)
        pass
    
    # If we're padded, truncate down to the target shape
    Dtime = numpy.fft.ifft2(Dcol, axes=(0,1))[:Dtime.shape[0], :Dtime.shape[1], :]
    
    Dtime = Dtime.reshape( (Dtime.shape[0], Dtime.shape[1] * Dtime.shape[2]), order='F')
    
    figure()
    imshow(librosa.logamplitude(Dtime), 
           aspect='auto', 
           interpolation='none', 
           origin='lower',
           vmax=0.0, vmin=-80.0)
    colorbar()
    
    # Get the y-axis labels
    nmels    = Dtime.shape[0]
    binfreqs = librosa.feature.mel_frequencies(n_mels=nmels)
    yticks(range(0, nmels+1, nmels/4), map(lambda x: '%dHz' % int(x), binfreqs[1:-1:(nmels/4)]))
    axis('tight')
    pass


def show_activation_magnitudes(A, m):
    
    (d2m, n) = A.shape
    
    d2 = d2m / m
    d  = d2  / 2
    
    # Combine real and imaginary components, compute energy
    energy = A[:(d2m/2)]**2 + A[(d2m/2):]**2
    
    # Aggregate per codeword
    Z = numpy.zeros((m, n))
    
    for i in range(m):
        Z[i,:] = numpy.sum(energy[ (i * d):((i+1)*d), :], axis=0) ** 0.5
        pass
    
    figure()
    imshow(Z, aspect='auto', interpolation='none'), colorbar()
    xlabel('Data')
    yticks(range(m), range(1, m+1))
    ylabel('Codewords')
    
    pass


def spatial_activations(A, W, M, pad=False):
    
    # Convert to complex
    AC = CDL.real2_to_complex(A)
    n  = AC.shape[1]
    
    if pad:
        AC = AC.reshape( (AC.shape[0] / (2 * W * M), 2 * W, M, n), order='F')
        Aspace   = numpy.fft.ifft2(AC, axes=(0,1)).real[:AC.shape[0]/2, :AC.shape[0]/2, :, :]
        pass
    else:
        AC = AC.reshape( (AC.shape[0] / (W * M), W, M, n), order='F')
        Aspace   = numpy.fft.ifft2(AC, axes=(0,1)).real
        pass
    
    return Aspace


def show_activation(Aspace):
    m = Aspace.shape[2]
    w = numpy.floor(m**0.5)
    
    figure()
    for i in range(m):
        subplot(w, numpy.ceil(float(m) / w), i+1)
        imshow(Aspace[:,:,i], aspect='auto', origin='lower', interpolation='none')
        title('Activation %2d' % (i+1))
        axis('off')
        pass
    pass


def suppress_codewords(A, m, s):
    Aout      = CDL.real2_to_complex(A)
    
    (dm, n)   = Aout.shape
    d         = dm / m
    
    v         = numpy.arange(d).astype(int)
    
    for i in s:
        Aout[d * i + v, :] = 0.0
        pass
        
    return CDL.complex_to_real2(Aout)


def show_admm_diagnostics(__DIAG, G, label):
    if type(__DIAG) is list:
        for d in __DIAG:
            show_admm_diagnostics(d, G, label)
            pass
        return
    time_range = numpy.linspace(0, __DIAG['num_steps'], len(__DIAG['err_primal']))
    semilogy(time_range, 1e-10 + __DIAG['err_primal'] / __DIAG['eps_primal'], 'r')
    semilogy(time_range, 1e-10 + __DIAG['err_dual'] / __DIAG['eps_dual'], 'b')
    semilogy(time_range, 1e-10 + __DIAG['rho'], 'm:')
    plot([time_range[0], time_range[-1]], [1, 1], 'k--')
    legend(['Primal', 'Dual', 'rho', 'Convergence'])
    title('[%s] Convergence' % label)
    xlabel('ADMM iterations')
    pass


def show_reconstruction(X, D, A, W, pad=False):
    figure()
    
    show_2d_basis(CDL.columns_to_diags(X), sample_width=W, pad=pad)
    title('X')
    
    show_2d_basis(CDL.columns_to_diags(D * A), sample_width=W, pad=pad)
    title('Reconstruction: D * A')
    
    show_2d_basis(CDL.columns_to_diags(D * A - X), sample_width=W, pad=pad)
    title('Reconstruction error: D * A - X')
    pass


def show_codebook(D, W, M, pad=False):
    show_2d_basis(D, sample_width=W, pad=pad)
    if pad:
        stem(numpy.arange(-0.5, M * W - 0.5, W), [X.shape[0] / (8 * W)] * M, 'w')
        axis([0, M * W, 0, X.shape[0] / (8 * W) - 1])
    else:
        stem(numpy.arange(-0.5, M * W - 0.5, W), [X.shape[0] / (2 * W)] * M, 'w')
        axis([0, M * W, 0, X.shape[0] / (2 * W) - 1])
        pass
    title('Dictionary - D [%d codewords, %d frames each]' % (M, W))
    xticks(range(W/2, M * W, W), range(1, M+1))
    xlabel('Codeword #')
    pass
```
